## 이미지 목록 불러오기

```python
ori_img_list = []

for i in range(start_read_point, max_file_name+1):
    if len(ori_img_list) == read_file_size:
        break


    file_name = ori_img_dir_name + "T_" + ("{:05d}".format(i)) + "_19_normcore_M.jpg"

    if not os.path.isfile(file_name): # 해당 파일명이 디렉토리에 존재하는지 체크
        continue

    # 파일이 존재하면
    ori_img_file = cv2.imread(file_name)

    if ori_img_file is None: # 파일 불러오기 실패
        print(file_name + " 파일 불러오기 실패")
        break
    else: # 파일 불러오기 성공
        end_read_point = i


    ori_img_list.append(ori_img_file)

    if (len(ori_img_list)%100) == 0: # 진행도 확인용
        print(str(len(ori_img_list)) + "개 이미지 파일 읽어오기 완료")


# 다 끝나면
print()
print(str(end_read_point) + ".jpg까지 파일 읽어오기 완료")
```

## json 목록 불러오기

```python
ori_json_list = []

for i in range(start_read_point, max_file_name+1):
    if (len(ori_json_list)) == read_file_size: # 최대 읽을 파일 개수까지 읽은 경우
        break

    if not os.path.isfile(ori_img_dir_name + "T_" + ("{:05d}".format(i)) + "_19_normcore_M.jpg"): # 해당 파일명이 존재하는지 체크
        continue


    os.chdir(ori_json_dir_name[:-1])
    file_names = os.listdir()
    json_names = []

    for name in file_names:
        if name.startswith("T_" + ("{:05d}".format(i)) + "_19_normcore_M"):
            json_names.append(name)

    file_name = json_names[0]

    # 파일이 존재하면
    with open(file_name, "r", encoding="UTF-8") as loadfile:
        ori_json_file = json.load(loadfile)

    if ori_json_file is None: # 파일 불러오기 실패
        print(file_name + " 파일 불러오기 실패")
        break
    else: # 파일 불러오기 성공
        end_read_point = i

    ori_json_list.append(ori_json_file)

    if (len(ori_json_list)%100) == 0: # 진행도 확인용
        print(str(len(ori_json_list)) + "개 json 파일 읽어오기 완료")


# 다 끝나면
print()
print(str(end_read_point) + ".json까지 파일 읽어오기 완료")
```

## 디렉토리에 저장

```python
os.chdir('/home/ubuntu/MachineLearningProject/02. Fashion Type Classification Model')
```

```python
file_num = start_save_point - 1 # 밑에서 +1을 하고 시작하므로

for ori_file in ori_file_list:

    file_num = file_num + 1

    if (file_num%100) == 0: # 진행도 체크용
        print(str(file_num) + "번째 파일 작업중...")

    img_file_path = proc_img_dir_name + str(file_num) + ".jpg"
    json_file_path = proc_json_dir_name + str(file_num) + ".json"


    # 이미지 파일 저장
    # flip_img = cv2.flip(ori_file[0], 1) # 1은 좌우반전, 0은 상하반전

    success = cv2.imwrite(img_file_path, ori_file[0])
    if not success:
        print(str(file_num) + "번째 이미지 파일 저장 실패!")
        break


    # json 파일 저장
    new_json = {}
    new_json['Img_Height'] = len(ori_file[0])
    new_json['Img_Width'] = len(ori_file[0][0])
    new_json['Labeling'] = ori_file[1]['item']['style']

    with open(json_file_path, 'w') as outfile:
        json.dump(new_json, outfile)

    success = os.path.isfile(json_file_path)
    if not success:
        print(str(file_num) + "번째 json 파일 저장 실패!")
        break


    # 무사히 저장이 완료
    end_save_point = file_num



print()
print(str(end_save_point) + "까지 파일 저장 성공!!")
```
